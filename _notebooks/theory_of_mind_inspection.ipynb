````python
from negobench.agents.agents import Agent

def inspect(game: Game, inspection_prompt):
    game = copy.deepcopy(game)
    for iteration, state in enumerate(game.game_state[1:-1]):
        current_iteration = state['current_iteration']
        # turn = state['turn']
        print("Current Iteration : {}".format(current_iteration))

        # mutate plyaer state
        for turn, ps in enumerate(state['player_state']):

            last_convo = ps['conversation'][current_iteration//2]
            print("\nTURN : {}".format(turn))
            print(last_convo['content'][:100],"...")
            if last_convo['role'] != 'user':
                state['player_state'][turn]['conversation'].append({'role':'user', 'content': inspection_prompt})
            else:
                state['player_state'][turn]['conversation'][-1]['content'] += inspection_prompt 

            agent = Agent.from_dict(ps)
            response = agent.think()
            print('\n')
            print(response)
            # print('\n')

inspect(game_states[3], inspection_prompt)

````
